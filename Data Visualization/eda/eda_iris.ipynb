# Memvisualisasikan data

```python
plt.title('Rata rata speciest')
sns.countplot(iris['Species'])
```

Insight:

data spesies bunga memiliki rata rata sama (50 data tiap spesies)




```python
plt.figure(figsize=(10, 6))
plt.title('Hubungan Spesies Berdasarkan Panjang dan Lebar Sepal')
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=iris, s=50)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Hubungan Spesies Berdasarkan Panjang dan Lebar Petal')
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=iris, s=50)
plt.show()
```

```python
sns.pairplot(iris,hue='Species')
```

```python
corr_matrix = iris.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
iris.groupby('Species').agg(['mean', 'median'])
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))
sns.boxplot( y="PetalWidthCm", x= "Species", data=iris, orient='v' , ax=axes[0, 0])
sns.boxplot( y="PetalLengthCm", x= "Species", data=iris, orient='v' , ax=axes[0, 1])
sns.boxplot( y="SepalLengthCm", x= "Species", data=iris, orient='v' , ax=axes[1, 0])
sns.boxplot( y="SepalWidthCm", x= "Species", data=iris, orient='v' , ax=axes[1, 1])
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes[0, 0].set_title("Sepal Length")
axes[0, 0].hist(iris['SepalLengthCm'], bins=7)

axes[0, 1].set_title("Sepal Width")
axes[0, 1].hist(iris['SepalWidthCm'], bins=5)

axes[1, 0].set_title("Petal Length")
axes[1, 0].hist(iris['PetalLengthCm'], bins=6)

axes[1, 1].set_title("Petal Width")
axes[1, 1].hist(iris['PetalWidthCm'], bins=6)

plt.tight_layout()
plt.show()

```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,10))
sns.violinplot(  y="PetalWidthCm", x= "Species", data=iris, orient='v' , ax=axes[0, 0],inner='quartile')
sns.violinplot( y="PetalLengthCm", x= "Species", data=iris, orient='v' , ax=axes[0, 1],inner='quartile')
sns.violinplot(y="SepalLengthCm", x= "Species", data=iris, orient='v' , ax=axes[1, 0],inner='quartile')
sns.violinplot( y="SepalWidthCm", x= "Species", data=iris, orient='v' , ax=axes[1, 1],inner='quartile')
plt.show()
```

```python
# Sepal Length
sns.displot(iris, x="SepalLengthCm", hue="Species", kind="kde", height=5)
plt.legend(title='Species')
plt.title('Distribution of Sepal Length by Species')
plt.show()

# Sepal Width
sns.displot(iris, x="SepalWidthCm", hue="Species", kind="kde", height=5)
plt.legend(title='Species')
plt.title('Distribution of Sepal Width by Species')
plt.show()

# Petal Length
sns.displot(iris, x="PetalLengthCm", hue="Species", kind="kde", height=5)
plt.legend(title='Species')
plt.title('Distribution of Petal Length by Species')
plt.show()

# Petal Width
sns.displot(iris, x="PetalWidthCm", hue="Species", kind="kde", height=5)
plt.legend(title='Species')
plt.title('Distribution of Petal Width by Species')
plt.show()
```
